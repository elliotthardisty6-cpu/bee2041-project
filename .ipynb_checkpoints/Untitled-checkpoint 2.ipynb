Do Premier League clubs get value for money?

Money has become an extremely important aspect of football, especially in the Premier League, where player wages are so high. Some clubs are spending huge amounts each season, with Manchester City reportedly paying over £350 million in wages during the 2023/2024 season. That figure alone is more than the total income of many clubs outside the top division, which suggests a huge imbalance.

This made me consider whether spending that level of money actually guarantees success or if it's sometimes a waste. It seems obvious that better players would cost a greater amount, but I wasn't completely sure how strong the relationship would be when you look at the data.

In this post, I look at and analyse data from the 2023/2024 Premier League season to see whether higher wage bills are linked to better performance, measured by league standings and points. I also look at which teams seemed to get the most value for money, and which teams didn’t do as well as they should have even with high expenditure.

To do this, I used a regression model to see if there was a link between performance data from the website Fbref and estimated wage data. I then looked at how each individual club performed compared to the predicted results from the model. This gave a clearer idea of the team’s efficiency rather than just spending.

Does Spending More Actually Lead to Better Results?

To formally test whether wage spending predicts league points, I used an Ordinary Least Squares (OLS) regression model. In simple terms, this method fits a straight line through the data to show the overall trend between two variables, which in this case is wage spending and league points.

Each team is plotted as a point on a scatter graph, and the regression line represents the average relationship between wages and performance. The slope of the line tells us how many extra points a team is expected to gain for additional £1 million spent.

What struck me as compelling here is that the model doesn't just show whether there is a relationship, but also how strong it is. Teams above the line performed better than expected given their wages, while teams below the line are underperformed.

```python
from IPython.display import Image
Image('figures/01_wages_vs_points.png')
```

```python
import statsmodels.formula.api as smf
import pandas as pd
 
df = pd.read_csv('data/cleaned/pl_with_residuals.csv')
model = smf.ols('Pts ~ annual_wages_millions', data=df).fit()
 
results = pd.DataFrame({
    'Coefficient': model.params,
    'Std Error':   model.bse,
    'P-value':     model.pvalues
}).round(3)
print(results)
print(f'\nR-squared: {model.rsquared:.3f}')
```

What Do the Results Tell Us?

The regression produces three key numbers worth discussing.

Firstly, the coefficient on annual wages tells that for each additional £1 million in wages, a club earns roughly 0.1854 extra points on average. At first that sounds relatively small, but when you think about the sale of spending in the league, it adds up quickly. For example, a £100 million difference in wages would be associated with around 18-19 extra points over a season, which is a large gap in the Premier League.

The p-value for this estimate is p<0.001, which means the relationship is statistically significant at any standard level. In simple terms, this means it's extremely unlikely that this result is just due to random chance. Therefore, there is very strong evidence that higher wage spending is linked to better performance. 

If we look at the R-squared value of 0.655, it suggests that about around 65.5% of the variation in points across teams can be explained by differences in wage spending alone. This value is significant, showing that money does play a really important role in determining where teams finish in the table. 

That said, it’s not everything. Around 35% of the variation is still left unexplained by the model, implying that other factors also matter. For example, tactics, injuries, squad depth and management decisions probably have an impact as well.

Overall, the evidence clearly supports the idea that money matters in the Premier League. Teams that spend more on wages tend to perform better, and the relationship is both statistically significant and strong.

Which Teams Got Value for Money?

Even though the regression shows a clear link between wages and performance, it doesn't explain everything. Some teams perform better or worse than the model predicts, which is where the idea of "value for money" becomes useful.

These differences are captured by something called residuals. A positive residual means a team performed better than expected based on their wages, while a negative one means they underperformed.

I found this part of the analysis the most interesting, as it highlights which clubs are genuinely efficient rather than just wealthy, and the result could be considered surprising.
